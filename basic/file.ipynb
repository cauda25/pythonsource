```python
# score.txt을 읽어 온 후 합계와 평균 구한 후 출력
f = open("./score.txt","r",encoding="utf-8")
scores = []
for num in f:
    scores.append(int(num))

print(f"평균 : {sum(scores) / len(scores)}")
   

```

```python
# sample.txt 읽은 후 합계와 평균 구하기
# result.txt 작성 (합계, 평균 기록)
sample = []
f = open("./sample.txt","r",encoding="utf-8")
for num in f:
    sample.append(int(num))

print(f"합계 : {sum(sample)}")
print(f"평균 : {sum(sample) / len(sample)}")


f1 = open("./result.txt","w",encoding="utf-8")
f1.write(f"합계 : {sum(sample)}\n")
f1.write(f"평균 : {sum(sample) / len(sample)}")
```

```python
# info.txt 읽기
f = open("./info.txt","r",encoding="utf-8")
# BMI 지수 계산 후 화면에 출력
# [출력결과]
# 아파 ,97 ,192
# 이름 : 아파
# 몸무게 : 97
# 키 : 192
# BMI : (bmi = weigth / (height / 100) ** 2)
# 결과 : 
for info in f:
    name, weight, height = info.strip().split(" ,")
    # print(name, weight, height)
    bmi = int(weight) / (int(height) / 100) ** 2

    if 25 <= bmi:
        result = "과체중"
    elif 18.5 <= bmi:
        result = "정상체중"
    else:
        result = "저체중"

    print(name, weight, height)
    print(f"이름 : {name}") 
    print(f"몸무게 : {weight}") 
    print(f"키 : {height}") 
    print(f"BMI : {bmi}") 
    print(f"결과 : {result}\n") 

     
```

```python

while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료"))
    if no == 1:
        # origin 읽고 암호화 후 encrytion.txt로 저장
        f = open("./origin.txt","r",encoding="utf-8")
        content = f.read()

        f1 = open("./encrytion.txt","w",encoding="utf-8")
        for c in content:
            f1.write(chr(ord(c) + 100))
    elif no == 2:
        # encrytion 을 읽고 원래 문자대로 화면 출력
        f2 = open("./encrytion.txt","r",encoding="utf-8")
        content = f2.read()
        for i in range(len(content)):
            print(chr(ord(content[i]) - 100), end="")
    else:
        break
```

```python
# close() 자동으로 호출해줌

with open("./origin.txt","r",encoding="utf-8") as f1:
    print(f1.read())
```
