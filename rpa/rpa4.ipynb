```python
# 네이버 이미지 다운
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
from urllib.request import urlretrieve

driver = set_chrome_driver()
driver.maximize_window()


driver.get("https://www.naver.com/")
time.sleep(1)

element = driver.find_element(By.ID, "query")
element.send_keys("슈화")
element.send_keys(Keys.ENTER)
time.sleep(2)

# 이미지 메뉴 클릭
driver.find_element(By.XPATH, '//*[@id="lnb"]/div[1]/div/div[1]/div/div[1]/div[1]/a').click()
time.sleep(2)
# 스크롤 스크립트
prev_height = driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

while True:
    # 스크롤 이동
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

    time.sleep(2)

    # 스크롤이 진행된 후 문서 높이
    curr_height = driver.execute_script("return scrollTo(0,document.body.scrollHeight)")

    if curr_height == prev_height:
        break

    prev_height = curr_height

# 맨 처음으로 가기
driver.execute_script("window.scrollTo(0,0)")

time.sleep(3)

# 이미지들 가져오기
images = driver.find_elements(By.CSS_SELECTOR,".thumb .link_thumb")
time.sleep(3)

count = 1
for img in images:
    # print(img.get_attribute('src'))
    img.click()
    time.sleep(2)

    # 오른쪽에 뜬 큰 이미지 저장
    large_img =  driver.find_element(By.CSS_SELECTOR, ".viewer_image img").get_attribute('src')

    urlretrieve(large_img,"./download/"+str(count)+".png")
    count += 1

time.sleep(3)

driver.quit() 
```

```python
# xlsx 저장장
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from openpyxl import Workbook
from openpyxl.drawing.image import Image
import requests
from fake_useragent import UserAgent
from io import BytesIO


wb = Workbook()
ws = wb.active
ws.column_dimensions["B"].width = 70
ws.column_dimensions["C"].width = 20
ws.append(['번호','제품명','가격','이미지'])

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(2)

driver.find_element(By.CSS_SELECTOR, "#dlMaker_simple > dd > div.spec_opt_view > button.btn_spec_view.btn_view_more").click()
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="searchMaker1452"]').click()
time.sleep(3)

cur_page,target_crawl_num = 1,5
idx =1

while cur_page <= target_crawl_num:
    print(f"현재 페이지 : {cur_page}")

    # 데이터 수집
    # 제품명, 가격(첫번째), img src
    product_list = driver.find_elements(By.CSS_SELECTOR, ".main_prodlist ul.product_list li.prod_item")
    for product in product_list:
        title = product.find_element(By.CSS_SELECTOR, ".prod_name a").text
        price = product.find_element(By.CSS_SELECTOR, "p.price_sect a strong").text
        img_src = product.find_element(By.CSS_SELECTOR, ".thumb_image img")

        if img_src.get_attribute("data-original"):
            src = img_src.get_attribute("data-original")
        else:
            src = img_src.get_attribute('src')

        if "https:" not in src:
            src = "https:"+ src
        
        print(idx,title,price,src)
        # 번호 제품명 가격 이미지
        ws.append([idx,title,price])
        idx += 1

        # 이미지 가져와서 엑셀 저장
        res = requests.get(src,headers={"user-agent":UserAgent().chrome})
        save_img = BytesIO(res.content) # content : byte 로 처리

        img = Image(save_img)
        img.width=30
        img.height=20
        ws.add_image(img,"D"+str(idx))

    
    
    cur_page += 1

    if cur_page > target_crawl_num:
        break
    # #productListArea > div.prod_num_nav > div > div > a.num.now_on
    # div.number_wrap > a:nth-child(2)
    # div.number_wrap > a:nth-child(3)
    # div.number_wrap > a:nth-child(4)
    # div.number_wrap > a:nth-child(5)
    driver.find_element(By.CSS_SELECTOR, f"div.number_wrap > a:nth-child({cur_page})").click()
    time.sleep(3)


wb.save("./danawa.xlsx")
driver.quit()

```

```python
# db 저장
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import sqlite3
import re

conn = sqlite3.connect("./danawa.db",isolation_level=None)
cursor = conn.cursor()

driver = webdriver.Chrome()
driver.maximize_window()

driver.get("https://prod.danawa.com/list/?cate=112758&15main_11_02=")
time.sleep(2)

driver.find_element(By.CSS_SELECTOR, "#dlMaker_simple > dd > div.spec_opt_view > button.btn_spec_view.btn_view_more").click()
time.sleep(1)
driver.find_element(By.XPATH, '//*[@id="searchMaker1452"]').click()
time.sleep(3)

cur_page,target_crawl_num = 1,5
idx =1

while cur_page <= target_crawl_num:
    print(f"현재 페이지 : {cur_page}")

    # 데이터 수집
    # 제품명, 가격(첫번째), img src
    product_list = driver.find_elements(By.CSS_SELECTOR, ".main_prodlist ul.product_list li.prod_item")
    for product in product_list:
        title = product.find_element(By.CSS_SELECTOR, "p.prod_name a").text
        price = product.find_element(By.CSS_SELECTOR, "p.price_sect a strong").text
        img_src = product.find_element(By.CSS_SELECTOR, ".thumb_image a.thumb_link img")

        if img_src.get_attribute("data-original"):
            src = img_src.get_attribute("data-original")
        else:
            src = img_src.get_attribute('src')

        if "http:" not in src:
            src = "http:"+src
        
        print(idx,title,price,src)
        # db 저장
        price = re.sub(",","",price)
        sql = "INSERT INTO appl(title, price, url) VALUES(?, ?, ?)"
        cursor.execute(sql, (title, price, src))
        
        idx += 1

    
    
    cur_page += 1

    if cur_page > target_crawl_num:
        break
    # #productListArea > div.prod_num_nav > div > div > a.num.now_on
    # div.number_wrap > a:nth-child(2)
    # div.number_wrap > a:nth-child(3)
    # div.number_wrap > a:nth-child(4)
    # div.number_wrap > a:nth-child(5)
    driver.find_element(By.CSS_SELECTOR, f"div.number_wrap > a:nth-child({cur_page})").click()
    time.sleep(3)


driver.quit()

```
